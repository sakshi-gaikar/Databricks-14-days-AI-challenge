
DAY-2 Apache Spark Fundamentals

```python
events=spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv",header=True,inferSchema=True)
```

```python
#Basic Operations: select , filter , groupby , orderby
events.select("event_type","product_id","price").show(10)
```

```python
events.filter("price>300").show(5)
```

```python
events.groupBy("event_type").count().show()
```

```python
top_brands=events.groupBy("brand").count().orderBy("count",ascending=False)
top_brands.show(5)
```
